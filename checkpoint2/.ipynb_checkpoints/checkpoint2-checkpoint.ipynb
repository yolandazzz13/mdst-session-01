<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python

X = df[['TOTAL_REVENUE','INSTRUCTION_EXPENDITURE','YEAR']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.3, random_state=0)
```

<h2> Prediction </h2>

ML Models Resource: https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a

Chosen ML task: **<#REGRESSION>**

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# FOR CLASSIFICATION ONLY:
#from sklearn.metrics import plot_confusion_matrix

#plot_confusion_matrix(model, X_test, y_test,
   #                      cmap=plt.cm.Blues)
```

```python
# FOR REGRESSION ONLY: (pick a single column to visualize results)

# Results from this graph _should not_ be used as a part of your results -- it is just here to help with intuition. 
# Instead, look at the error values and individual intercepts.

#INSTRUCTION_EXPENDITURE
#TOTAL_REVENUE
col_name = 'YEAR'
col_index = X_train.columns.get_loc(col_name)

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
plt.scatter(X_test[col_name], model.predict(X_test), color = "blue")

new_x = np.linspace(X_train[col_name].min(),X_train[col_name].max(),200)
intercept = model.predict([X_train.sort_values(col_name).iloc[0]]) - X_train[col_name].min()*model.coef_[col_index]
plt.plot(new_x, intercept+new_x*model.coef_[col_index])

plt.legend(['controlled model','true training','predicted training','predicted testing'])
plt.xlabel(col_name)
plt.ylabel('4th grade math score')
```

<h2> Summary </h2>

There is a weak negative correlation between total revenue and average 4th grade math scores. The correlation is extremely weak that there's hardly any slope in the controlled model. 

Similarly, there is a weak positive correlation between instruction expenditure and average 4th grade math scores. The correlation is also extremely weak.

More data tend to gather in the lower total revenue/instruction expenditure area. This is due to the financial situations of different state. The pattern is quite similar to that of income, where most data gather in the lower portion, with a few lying on the far right end, representing extremely high income. With more data, the model makes more reliable prediction in the lower portion. 

As we can see from the graph, majority of the data lie between 230 and 240. After computing the conditional means over different years, the dispersion are cancelled out and they converge to a steady value under different total revenue/ instruction expenditure. So the correlation is weak.

However, when considering year as the independent variable, there's an obvious trend that student's 4th grade math scores increase as time proceeds. Therefore there's a positive corelation between the two.
